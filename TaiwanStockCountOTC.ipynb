```python
# To Read Taiwan Tourism Bureau data and make it into HTML
# Taiwan Stock Data1 https://bsr.twse.com.tw/bshtm/
# Filename: 2603_20200818.csv
# Taiwan Stock Data2 https://www.tpex.org.tw/web/stock/aftertrading/broker_trading/brokerBS.php
# Load basic libraries
import pandas as pd
import numpy as np
```

```python
# Check first 5 rows in a file
# from tabula import 
df = pd.read_csv('3218_20200915.CSV',encoding='utf-8',skiprows=2)

# Notice datatypes changed in some columns.
# Check first few rows.
df.head()
```

```python
# Drop some columns.
df = df.drop(df.columns[[5]], axis=1) 
df.head()
```

```python
# Split fields into 2 files.
# Copy files
df1 = df
df2 = df
df1 = df1.drop(df.columns[[5,6,7,8,9]], axis=1) 
df1.head()
```

```python
# Do it again
df2 = df2.drop(df2.columns[[0,1,2,3,4]], axis=1) 
df2.head()
```

```python
# check datatypes
df2.dtypes
```

```python
# Change column names
df2.columns = ['序號','券商', '價格', '買進股數', '賣出股數']
df2.head()
```

```python
# join columns
df3 = pd.concat([df1, df2])
df3.head()
```

```python
# check last few rows.
df3.tail()
```

```python
# Check if we need to delete the last row?
# Check if there is a Nan?
# delete last row
df3.dropna(subset = ["買進股數"], inplace=True)
df3.tail()
```

```python
# Sort data
df4 = df3.sort_values(by=['序號'])
df4.head()
```

```python
# reindex
#df4.reset_index
df4 = df4.reset_index()
df4 = df4.drop(df4.columns[[0,1]], axis=1) 
df4.head()
```

```python
# delete the price column
df8 = df4.drop(df4.columns[[1]], axis=1) 
df8.head()
```

```python
# Check last item
df8.tail(3)
```

```python
# Check if we need to delete the last row?
# delete last row
df3.drop(df3.tail(1).index,inplace=True)
df3.tail()
```

```python
# check datatypes
df8.dtypes
```

```python
# Check data type again
# Change datatypes
df8["買進股數"] = df8["買進股數"].str.replace(',', '')
df8["買進股數"] = df8["買進股數"].astype(str).astype(int)
df8["賣出股數"] = df8["賣出股數"].str.replace(',', '')
df8["賣出股數"] = df8["賣出股數"].astype(str).astype(int)
df8.dtypes
```

```python
# Check if datatypes are integers?
df8.head()
```

```python
# Sum of 2 columns
df5 = df8.groupby(['券商']).agg({'買進股數':'sum','賣出股數':'sum'})
df5.head()
```

```python
# Sub to get values and sort.
df5['差異']=df5['買進股數']-df5['賣出股數']
df5.head()
```

```python
# Sort by Value-差異
df6 = df5.sort_values(by=['差異'])

# return first 40 rows
# 顯示賣出量最大40間券商
df6.head(40)
```

```python
# Reverse sort
df7 = df5.sort_values(by=['差異'],ascending=False)

# return first 40 rows
# 顯示買進量最大40間券商
df7.head(40)
```
