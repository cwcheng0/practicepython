```
# KNeighborsClassifier
# 2019-1-26
# Source https://www.smwenku.com/a/5c2206e8bd9eee16b4a76907
# Source2 https://ithelp.ithome.com.tw/articles/10199904
# Source3 https://medium.com/@yehjames/%E8%B3%87%E6%96%99%E5%88%86%E6%9E%90-%E6%A9%9F%E5%99%A8%E5%AD%B8%E7%BF%92-%E7%AC%AC2-1%E8%AC%9B-%E5%A6%82%E4%BD%95%E7%8D%B2%E5%8F%96%E8%B3%87%E6%96%99-sklearn%E5%85%A7%E5%BB%BA%E8%B3%87%E6%96%99%E9%9B%86-baa8f027ed7b
# Source4 https://machine-learning-python.kspax.io/datasets/ex3_the_iris_dataset

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn import datasets
from sklearn.model_selection import train_test_split #Split arrays or matrices into random train and test subsets
import matplotlib.pyplot as plt

#load進iris的資料庫
iris = datasets.load_iris()  

features = iris.feature_names
X = iris.data  
y = iris.target

#test_size預設是0.25
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)  

k_range = range(1,10)  
k_scores = []
for k in k_range:
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X_train, y_train)
    y_result = neigh.predict(X_test) 
    k_scores.append(accuracy_score(y_test, y_result))
    print('neighbors: ', k, 'score: ', accuracy_score(y_test, y_result))
    
   
# Draw the picture
plt.plot(k_range,k_scores)
plt.xlabel('Value of K')
plt.ylabel('Accuracy')
plt.show()
```

```
# Concatenating multiple feature extraction methods
from sklearn.pipeline import Pipeline, FeatureUnion
#from sklearn.grid_search import GridSearchCV <- Fix an error
from sklearn.model_selection import GridSearchCV

from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest

iris = load_iris()

X, y = iris.data, iris.target

# This dataset is way to high-dimensional. Better do PCA:
pca = PCA(n_components=2)

# Maybe some original features where good, too?
selection = SelectKBest(k=1)
```

```
# Build estimator from PCA and Univariate selection:
combined_features = FeatureUnion([("pca", pca), ("univ_select", selection)])

# Use combined features to transform dataset:
X_features = combined_features.fit(X, y).transform(X)
```

```
# SVM
svm = SVC(kernel="linear")

# Do grid search over k, n_components and C:
pipeline = Pipeline([("features", combined_features), ("svm", svm)])

param_grid = dict(features__pca__n_components=[1, 2, 3],
                  features__univ_select__k=[1, 2],
                  svm__C=[0.1, 1, 10])

grid_search = GridSearchCV(pipeline, param_grid=param_grid, verbose=10)
grid_search.fit(X, y)
print(grid_search.best_estimator_)
```
