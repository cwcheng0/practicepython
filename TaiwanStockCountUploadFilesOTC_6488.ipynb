```
# Check first 5 rows in a file
# OTC uses a different method to read Big5: Big5 ONLY
# df = pd.read_csv(fn ,encoding='big5hkscs',skiprows=2)

# OTC uses a different method to read Big5: UTF-8 ONLY
df = pd.read_csv(fn,encoding='utf-8',skiprows=2)
df.head()
```

```
# Drop a column.
df = df.drop(df.columns[[5]], axis=1) 
df.head()
```

```
# Split fields into 2 files.
# Copy files
df1 = df
df2 = df
df1 = df1.drop(df.columns[[5,6,7,8,9]], axis=1) 
df1.head()
```

```
# Do it again
df2 = df2.drop(df2.columns[[0,1,2,3,4]], axis=1) 
df2.head()
```

```
# check datatypes
df2.dtypes
```

```
# Change column names
df2.columns = ['序號','券商', '價格', '買進股數', '賣出股數']
df2.head()
```

```
# cahnge datatypes
#df2['買進股數'] = df2['買進股數'].astype(int)
#df2.head()
```

```
# join columns
df3 = pd.concat([df1, df2])
df3.head()
```

```
# check last few rows.
df3.tail()
```

```
# Check if we need to delete the last row?
# delete last row
df3.drop(df3.tail(1).index,inplace=True)
df3.tail()
```

```
# Sort data
df4 = df3.sort_values(by=['序號'])
df4.head()
```

```
# reindex
#df4.reset_index
df4 = df4.reset_index()
df4 = df4.drop(df4.columns[[0,1]], axis=1) 
df4.head()
```

```
# delete the price column
df8 = df4.drop(df4.columns[[1]], axis=1) 
df8.head()
```

```
# check datatypes
df8.dtypes
```

```
# Check data type again
# Change datatypes
df8["買進股數"] = df8["買進股數"].str.replace(',', '')
df8["買進股數"] = df8["買進股數"].astype(str).astype(int)
df8["賣出股數"] = df8["賣出股數"].str.replace(',', '')
df8["賣出股數"] = df8["賣出股數"].astype(str).astype(int)
df8.dtypes
```

```
# Check again
df8.head()
```

```
# Sum of 2 columns
df5 = df8.groupby(['券商']).agg({'買進股數':'sum','賣出股數':'sum'})
df5.head()
```

```
# Sub to get values and sort.
df5['差異']=df5['買進股數']-df5['賣出股數']
df5.head()
```

```
# Sort by Value-差異
df6 = df5.sort_values(by=['差異'])

# Change a column name
df6.columns = ['買進股數', '賣出股數', '今日賣超']

# Add a comma into an integer column
df6['今日賣超'] = df6.apply(lambda x: "{:,}".format(x['今日賣超']), axis=1)

# return first 40 rows
# 顯示賣出量最大40間券商
df6.head(40)
```

```
# Reverse sort
df7 = df5.sort_values(by=['差異'],ascending=False)

# return first 40 Buying Power rows
# 顯示買進量最大30間券商
df7.columns = ['買進股數', '賣出股數', '今日買超']

# Add a comma into an integer column
df7['今日買超'] = df7.apply(lambda x: "{:,}".format(x['今日買超']), axis=1)

# return first 40 rows
# 顯示買進量最大40間券商
df7.head(40)
```
