```
# SKLearn Decision Tree
# SKLearn practice using Wine as an example
# source http://www.peixun.net/view/1281.html
# 2019-1-21
from scipy.stats import norm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

wine = load_wine()
```

```
wine
```

```
# Check target's classifications
wine.target
```

```
# List the array
pd.concat([pd.DataFrame(wine.data),pd.DataFrame(wine.target)],axis=1)
```

```
#Feature names
wine.feature_names
```

```
# Target names
wine.target_names
```

```
# Split and set complexity
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data,wine.target,test_size=0.3)
```

```
# check shape
Xtrain.shape
```

```
# Check data shape 
wine.data.shape
```

```
# Using entropy to calculate the accuracy 
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)
score
```

```
# Using gini to calculate the accuracy 
clf = tree.DecisionTreeClassifier(criterion="gini", max_depth=3)
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)
score
```

```
# import Graphics to make a picture
import graphviz

dot_data = tree.export_graphviz(clf)
```

```
feature_name = ['酒精','蘋果酸','灰','灰的鹼性','鎂','總酚','類黃酮','非黃烷類酚類','花青素','顏 色強度','色調','od280/od315稀釋葡萄酒','脯氨酸']

dot_data = tree.export_graphviz(clf
                                ,feature_names= feature_name 
                                ,class_names=["琴酒","雪莉","貝爾摩德"] 
                                ,filled=True,rounded=True)   
#feature_names特徵索引；filled：顏色；rounded：控制數方框形狀。
graph = graphviz.Source(dot_data)
graph
```
