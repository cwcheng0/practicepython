```
# Check first 5 rows in a file
# from tabula import 
df = pd.read_csv(r'/content/gdrive/My Drive/ContestData/FinanceData/2353_20200807-2.csv',encoding='utf-8')

# delete first 2 rows
# df.drop(1,0,inplace=True)
# df.drop(1,0,inplace=True)

df.head()
```

```
# Drop some columns
df = df.drop(df.columns[[5]], axis=1) 
df.head()
```

```
# Split fields
# Copy files
df1 = df
df2 = df
df1 = df1.drop(df.columns[[5,6,7,8,9]], axis=1) 
df1.head()
```

```
# Do it again
df2 = df2.drop(df2.columns[[0,1,2,3,4]], axis=1) 
df2.head()
```

```
# Change column names
df2.columns = ['序號', '券商', '價格', '買進股數', '賣出股數']
df2.head()
```

```
# join columns
df3 = pd.concat([df1, df2])
df3.head()
```

```
# tail
df3.tail()
```

```
# Sort data
df4 = df3.sort_values(by=['序號'])
df4.head()
```

```
# reindex
#df4.reset_index
df4= df4.reset_index()
df4 = df4.drop(df4.columns[[0,1]], axis=1) 
df4.head()
```

```
# delete price column
df4 = df4.drop(df4.columns[[1]], axis=1) 
df4.head()
```

```
# Sum of some columns
df5 = df4.groupby('券商').sum()
df5.head()
```

```
# Sub
df5['差異']=df5['買進股數']-df5['賣出股數']
df5.head()
```

```
# Sort by Value
df6 = df5.sort_values(by=['差異'])
```
